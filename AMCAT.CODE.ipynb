Relationship between specialization and gender

```python
plt.figure(figsize=(10, 6))
sns.countplot(x='Specialization', hue='Gender', data=Eda)

# Add title and labels
plt.xticks(rotation=90)
plt.title('type of sepcialization based on their gender')
plt.xlabel('Specialization')
plt.ylabel('Gender')

# Show the plot
plt.show()
```

Top 5 specializations opted by female

```python
filtered_data = Eda[Eda['Gender'].isin(['f'])]
grouped_counts = filtered_data.groupby(['Specialization', 'Gender']).size().reset_index(name='Count')

# Step 2: Sort by count in descending order and take the top 5 most frequent combinations
top_5 = grouped_counts.sort_values(by='Count', ascending=False).head(5)

# Step 3: Plotting a bar chart for the top-5 relationships
plt.figure(figsize=(10, 6))

# Create labels for the bar plot by combining 'Department' and 'JobCity'
labels = top_5.apply(lambda row: f"{row['Specialization']} - {row['Gender']}", axis=1)

# Plot a horizontal bar plot
plt.barh(labels, top_5['Count'], color='skyblue')

# Add labels and title
plt.xlabel('Count')
plt.ylabel('Specialization - Gender')
plt.title('Top-5 Relationships Between Specialization and Gender')

# Display the plot
plt.show()
```

```python
plt.figure(figsize=(10,6))
degree_salary = Eda.groupby('Degree')['Salary'].mean().sort_values(ascending=False)
sns.barplot(x=degree_salary.index, y=degree_salary.values, palette='coolwarm')
plt.xticks(rotation=90)
plt.title('Average Salary by Degree')
plt.xlabel('Degree')
plt.ylabel('Average Salary')
plt.show()

# ANOVA test to check if there is a significant difference in Salary by Degree
degrees = Eda['Degree'].unique()
degree_salary_groups = [Eda[Eda['Degree'] == degree]['Salary'] for degree in degrees]
anova_result = stats.f_oneway(*degree_salary_groups)

print(f"ANOVA F-statistic: {anova_result.statistic}")
print(f"P-Value: {anova_result.pvalue}")
```
